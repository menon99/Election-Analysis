```python
import pandas as pd
df_1989 = pd.read_csv("1989.csv")
df_1991 = pd.read_csv("1991.csv")
df_1996 = pd.read_csv("1996.csv")
df_1998 = pd.read_csv("1998.csv")
df_1999 = pd.read_csv("1999.csv")
df_2004 = pd.read_csv("2004.csv")
df_2009 = pd.read_csv("2009.csv")
df_2014 = pd.read_csv("2014.csv")
```

```python
def f1(state,year,df):
    
    t1 = df.loc[df["TERRITORY"] == state]
    const = list(t1["CONSTITUENCY"].unique())
    
    won = 0
    tv = 0
    tvw = 0
    total_c = 0
    tp = 0
    
    for i in const:
        
        tc = df.loc[df["CONSTITUENCY"] == i]
        ti = list(tc.index)
        tv += tc["TOTAL VALID VOTES"][ti[0]]
        tf = tc.loc[tc["SEX"] == 'F']
        
        if tf.shape[0] !=0:
            total_c += 1

        for j in list(tf.index):
            
            if tf["No."][j] == 1:
                won+=1
            
            tvw += tf["VOTES"][j]
            tp += 1
    
    if(tv == 0):
        perc = 0
    else:
        perc = round(tvw/tv,5)*100
    temp = pd.DataFrame(data = [[year,won,total_c,tp,perc]] ,columns = ['year','won','contested','total_participation','% of votes'])
    return temp
```

```python
def f2(state,year,df):
    
    t1 = df.loc[df["TERRITORY"] == state]
    const = list(t1["CONSTITUENCY"].unique())
    
    won = 0
    tv = 0
    tvw = 0
    total_c = 0
    tp = 0
    
    for i in const:
        
        tc = df.loc[df["CONSTITUENCY"] == i]
        ti = list(tc.index)
        tv += tc["TOTAL VOTES POLLED"][ti[0]]
        tf = tc.loc[tc["SEX"] == 'F']
        
        
        if tf.shape[0] !=0:
            total_c += 1

        for j in list(tf.index):
            
            if tf["No"][j] == 1:
                won+=1
            
            tvw += tf["TOTAL VOTES RECEIVED"][j]
            tp += 1
    
    if(tv == 0):
        perc = 0
    else:
        perc = round(tvw/tv,2)*100
    temp = pd.DataFrame(data = [[year,won,total_c,tp,perc]] ,columns = ['year','won','contested','total_participation','% of votes'])
    return temp
```

```python
def func(state,d):
    
    temp = pd.DataFrame(columns = ['year','won','contested','total_participation','% of votes'])
    
    t1 = f1(state,1989,df_1989)
    t2 = f1(state,1991,df_1991)
    t3 = f1(state,1996,df_1996)
    t4 = f1(state,1998,df_1998)
    t5 = f1(state,1999,df_1999)
    t6 = f2(state,2004,df_2004)
    
    temp = temp.append(t1)
    temp = temp.append(t2)
    temp = temp.append(t3)
    temp = temp.append(t4)
    temp = temp.append(t5)
    temp = temp.append(t6)
    
    print("ANALYSIS FOR",state,'\n')
    print(temp.to_string(index=False))
    d[state] = temp['total_participation'].sum()
    print('\n')
```

```python
states = list(df_1989["TERRITORY"].unique())
```

## statistics of women in each territory

### WON represents the number of women who won in that territory for that year

### CONTESTED represents the total number of constituencies that women stood in

### TOTAL PARTICIPATION represents the total women who contested (multiple women can stand in a single constituency)

### % OF VOTES is total votes polled by women across all constituencies / total votes across all constituencies(men + women)

```python
d = {}
for i in states:
    func(i,d)
```
