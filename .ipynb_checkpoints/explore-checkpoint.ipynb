```python
class candidate:
    
    name = ''
    votes = 0
    perc = 0
    
    def create(self,n,v,p):
        
        self.name = n
        self.votes = v
        self.perc = p

class const:
    state = ''
    c_name = ''
    year = ''
    c_list = []
   
    
    def __init__(self,s,c,yr):
        
        self.c_name = c
        self.state = s
        self.year = yr
    
    def add_rec(self,name,votes,perc):
        
        c = candidate()
        c.create(name,votes,perc)
        c_list.append(c)
```
